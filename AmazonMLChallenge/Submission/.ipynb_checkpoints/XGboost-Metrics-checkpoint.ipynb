### Data Preprocessing

> As we cannot remove null rows from test so we have to impute it, I am doing it before one Hot encoding.

```python
# keeping customer category asid
```

```python
#getting categorical column
def categorical_columns(df):
    cat_col=[]
    num_col=[]
    col = df.columns
    for i in col:
        if df[i].dtype=='object':
            cat_col.append(i)
        else:
            num_col.append(i)
    return cat_col, num_col
Train_cat_col,Train_num_col = categorical_columns(train_set)
Test_cat_col,Test_num_col = categorical_columns(test_set)
print(Train_cat_col, Test_cat_col, Test_num_col)
```

```python
train_set.isnull().sum(),test_set.isnull().sum()
```

```python
#test_set[:]
```

```python
from sklearn.impute import SimpleImputer
def Imputating_data(df,cat_columns,num_col):
    for i in cat_columns:
        imputer = SimpleImputer(missing_values = np.nan,  
                            strategy ='most_frequent')
        imputer = imputer.fit(df[cat_columns])
        df_cat_col = imputer.transform(df[cat_columns])
    df_cat_col = pd.DataFrame(df_cat_col)
    df_cat_col.columns=cat_columns
    #print(df_cat_col.head())
    for i in num_col:
        imputer = SimpleImputer(missing_values = np.nan,  
                        strategy ='median')
        imputer = imputer.fit(df[num_col])
        df_num_col = imputer.transform(df[num_col])
    df_num_col = pd.DataFrame(df_num_col)
    df_num_col.columns=num_col
        
    df = pd.concat([df_cat_col,df_num_col],axis=1)
    return df


test_set = Imputating_data(test_set,Test_cat_col,Test_num_col)    
```

```python
#Checking for nulls
test_set.isnull().sum()
```

> we imputed test set, for the numerical column i have used median as we skew data and mean is influenced by skewness do i chose median value. For categorical, i have used most_frequent values.

>Next, we will now work on converting categorical column to numerical using one Hot encoding.

```python
def oneHotEncoding(cat_columns, df):
    df1=df
    for i in cat_columns:
        onehot = pd.get_dummies(df[i],prefix=i)
        df = pd.concat([df,onehot],axis =1).drop([i],axis=1)
    return df
        
transformed_train_set = oneHotEncoding(Train_cat_col,train_set)
transformed_test_set = oneHotEncoding(Test_cat_col,test_set)
```

### Model 2:Trying SMOTE with Logistic Regression

> A technique similar to upsampling is to create synthetic samples. Here we will use imblearn’s SMOTE or Synthetic Minority Oversampling Technique. SMOTE uses a nearest neighbors algorithm to generate new and synthetic data we can use for training our model.

```python


Y_train = transformed_train_set['customer_category']
X_Train = transformed_train_set.drop(['customer_category'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X_Train, Y_train, test_size=0.25, random_state=27)
sm = SMOTE(random_state=27, ratio=1.0)
X_train, y_train = sm.fit_sample(X_train, y_train)
smote = LogisticRegression(solver='liblinear').fit(X_train, y_train)

smote_pred = smote.predict(X_test)

# Checking accuracy
print(accuracy_score(y_test, smote_pred))
    #0.9858

# f1 score
print(f1_score(y_test, smote_pred))
    #0.1846

print(recall_score(y_test, smote_pred))
```

```python
# SMote and Logistic Regression
transformed_test_set.drop(['customer_id','customer_category'], axis='columns', inplace=True)
transformed_test_set['customer_category']=smote.predict(transformed_test_set)
transformed_test_set['customer_id']=customer_id_test_df
transformed_test_set.to_csv("Submission_smote_Metric_Model.csv",columns=['customer_id','customer_category'],index=False)
# And on Submission it give 80.39 Score
```

### Model 4: Try Param Tuned XG Boost Classifier.

> the scale_pos_weight can be used to train a class-weighted or cost-sensitive version of XGBoost for imbalanced classification.

> scale_pos_weight = total_negative_examples / total_positive_examples

```python
from collections import Counter
# count examples in each class
counter = Counter(Y_train)
print(counter)
# estimate scale_pos_weight value
estimate = counter[0] / counter[1]
print(estimate)
```

```python
#param_grid = {'learning_rate': [0.1,0.01,1], 'n_estimators': 1000, 'seed':0, 'subsample': 0.8, 'colsample_bytree': 0.8, 'max_delta_step':1,
#             'objective': 'binary:logistic','scale_pos_weight': estimate,'max_depth': [3,5,7], 'min_child_weight': [1,3,5]}
# define model
model = XGBClassifier()
# define evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
cv_params = {'max_depth': [3,5,7], 'min_child_weight': [1,3,5]}
ind_params = {'learning_rate': 0.1, 'n_estimators': 1000, 'seed':0, 'subsample': 0.8, 'colsample_bytree': 0.8, 'max_delta_step':1,
             'objective': 'binary:logistic','scale_pos_weight': estimate}
optimized_GBM = GridSearchCV(XGBClassifier(**ind_params), 
                            cv_params, 
                             scoring = 'roc_auc', cv = 5, n_jobs = -1) 

# execute the grid search
optimized_GBM_Model = optimized_GBM.fit(X_Train, Y_train)

#And on Submission it give 86.65 Score
```

```python
print("Best: %f using %s" % (optimized_GBM_Model.best_score_, optimized_GBM_Model.best_params_))
```

```python
optimized_GBM_Model
transformed_test_set.drop(['customer_id','customer_category'], axis='columns', inplace=True)
transformed_test_set['customer_category']=optimized_GBM_Model.predict(transformed_test_set)
transformed_test_set['customer_id']=customer_id_test_df
transformed_test_set.to_csv("Submission_optimized_GBM_Model_Metric_Model.csv",columns=['customer_id','customer_category'],index=False)
# And on Submission it give 86.05 Score
```
